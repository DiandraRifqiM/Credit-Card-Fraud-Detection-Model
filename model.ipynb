Phase 3 Data Preprocessing

Data Splitting

```python
# print(Cols)
X = df[['Time', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10',
       'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20',
       'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount']]

y = df['Class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

Preprocessing

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
# X_train
```

Phase 4 Model Training & Evaluate

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logReg = LogisticRegression()
logReg.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report
yPred1 = logReg.predict(X_test)
score1 = classification_report(y_test, yPred1, output_dict=True)
# print(f"Logistic Regression Score\n{score1}")
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
n = 4
RfC = RandomForestClassifier(n_estimators= n, random_state=42)
RfC.fit(X_train, y_train)
```

```python
yPred2 = RfC.predict(X_test)
score2 = classification_report(y_test, yPred2, output_dict=True)
# print(f"Logistic Regression Score\n{score2}")
```

Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
GbC = GradientBoostingClassifier(n_estimators=n, random_state=42)
GbC.fit(X_train, y_train)
```

```python
yPred3 = GbC.predict(X_test)
score3 = classification_report(y_test, yPred3, output_dict=True)
# print(f"Logistic Regression Score\n{score3}")
```

Performance Summary Each Model

```python
Model1 = pd.DataFrame((score2))
Model1
plt.pie(Model1["accuracy"], labels=Model1.index, autopct='%1.1f%%', startangle=90, colors=['skyblue','lightgreen','salmon', 'cyan'], shadow=True)
plt.title("Logistic Regression Acc Score")
```
